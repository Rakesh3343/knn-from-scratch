# *KNN Algorithm from scratch*

K nearest neigbhour algorithm belongs to the family of Supervised learning. It is used to solve both regression and classification problems.

##### What happens in K-nearest neigbour algorithm? and How does it give us predictions?

K-nearest algorithm assumes that any data point will have similar properties of its nearest K points.

KNN computes the distance b/w our query point to every data point, It sorts the data points in the ascending order of their distances and picks first k number of neighbors.  
After picking first k neighbours(i.e first k data points) if it is a regression problem it computes mean,if it is a classification problem it does majority voting and gives us the result.
